### shuffle Images

```python
np.random.shuffle(input_image)
```

```python
img=[]
lab=[]

for x_values, label in input_image:
    img.append(x_values)
    lab.append(label)
```

```python
img = np.array(img)
```

```python
# normalization of image

img = img/255
```

### Split the Data Into Train & Test

```python
img_train, img_test, lab_train, lab_test = train_test_split(img, lab, random_state=42, test_size=0.2, stratify=y)
```

```python
lab_train = to_categorical(lab_train, num_classes=3)
lab_test = to_categorical(lab_test, num_classes=3)
```

### Create CNN Model

```python
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import GlobalAveragePooling2D

model = Sequential()
model.add(Conv2D(filters=32, kernel_size=(3,3), activation='relu', input_shape=(300,300,3)))
# model.add(BatchNormalization())
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu'))
# model.add(BatchNormalization())
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Conv2D(filters=128, kernel_size=(3,3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPool2D(pool_size=(2,2)))

model.add(GlobalAveragePooling2D())
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(3, activation='softmax'))

model.summary()
```

### Compile the Model

```python
from tensorflow.keras.optimizers import Adam

model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

### Train the Model 

```python
model.fit(img_train,lab_train, epochs=20, validation_split=0.2)
```

The model exhibits clear overfitting behavior. Training accuracy increases rapidly
and reaches nearly 100%, while validation accuracy remains significantly lower
(around 65‚Äì70%) and fluctuates across epochs. Additionally, validation loss does
not consistently decrease, indicating that the model memorizes training samples
but fails to generalize well due to the limited dataset size.

### Check Model Performance

```python
df=pd.DataFrame(model.history.history)
df
```

### Model Performance Plot

```python
plt.figure(figsize=(12,7))
plt.plot(df.loss, color='red', label='Train', marker='*')
plt.plot(df.val_loss, color='blue', label='Validation', marker='*')
plt.legend()
```

```python
# Predict on test image
pred = model.predict(img_test)
```

```python
y_pred = np.argmax(pred, axis=1)
```

```python
y_true = np.argmax(lab_test, axis=1)
```

```python
# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
```

```python
# Heatmap on leaf data
plt.figure(figsize=(8,8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
           xticklabels=['Bacterial Leaf Blight','Brown Spot','Leaf Smut'],
            yticklabels=['Bacterial Leaf Blight','Brown Spot','Leaf Smut'])
plt.xlabel('predicted lable')
plt.ylabel('true lable')
plt.title('confusion matrix')
plt.show()
```

```python
# Classification Report
print(classification_report(
    y_true,
    y_pred,
    target_names=['Bacterial Leaf Blight','Brown Spot','Leaf Smut']
))
```

```python
# Check Accuracy
test_accuracy = accuracy_score(y_true, y_pred)
print("Test Accuracy:", test_accuracy)
```

**Model Performance**
* The CNN model was evaluated using accuracy curves, confusion matrix, and classification metrics. The model achieved a test accuracy of 62.5%, successfully classifying three rice leaf diseases: Bacterial Leaf Blight, Brown Spot, and Leaf Smut. The confusion matrix shows that most samples were correctly classified, with some misclassification occurring between visually similar disease classes.

### Apply Data Augmentation & Retrain CNN

### Data Augmentation

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=25,
    zoom_range=0.2,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    brightness_range=[0.8, 1.2],
    validation_split=0.3  # 70% train, 30% temp
)

train_generator = datagen.flow_from_directory(
    r'C:\Users\Administrator\Documents\Rubixi internship\PRCP-1001-RiceLeaf\Data',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='training',
    seed = 42,
)

num_classes = train_generator.num_classes
```

```python
val_generator = datagen.flow_from_directory(
    r'C:\Users\Administrator\Documents\Rubixi internship\PRCP-1001-RiceLeaf\Data',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='validation',
    seed =42
)
```

```python
print(train_generator.samples, val_generator.samples)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)
```

### CNN Model with augmentation

```python
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=40,
    callbacks=[early_stop]
)
```

```python
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='VAlidation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
val_generator.reset()

y_true = val_generator.classes
y_pred = np.argmax(model.predict(val_generator), axis=1)
```

```python
print('confusion matrix:')
print(confusion_matrix(y_true, y_pred))

print("\n classification report:")
print(classification_report(y_true, y_pred))
```
